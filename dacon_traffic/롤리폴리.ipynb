```python
gps = pd.read_csv("서울시_주요_기차역+터미널_좌표.csv")
```

```python
#데이터 셋 버스터미널/기차역으로 분류
train = gps.loc[:5]
minal = gps.loc[6:]
```

```python
train
```

```python
minal.reset_index(drop=True, inplace=True)
```

```python
minal
```

```python
#구별로 구역을 나누어줄 파일 다운로드
geo_path = '02. skorea_municipalities_geo_simple.json'
geo_str = json.load(open(geo_path, encoding='utf-8'))
```

```python
#이쁘게 구 별로 나누어진 모습
map = folium.Map(location=[37.5502, 126.982], zoom_start=11,
                tiles='stamentoner')
map.choropleth(geo_data=geo_str, fill_color = 'skyblue')
map.save('index.html')
map
```

```python
#기차역 좌표 추가
for i in range(len(train)):
  x = train['위도'][i]
  y = train['경도'][i]
  folium.Marker([x, y], tooltip=train['이름'][i], popup=train['이름'][i]).add_to(map)
```

```python
for i in range(len(minal)):
  x = minal['위도'][i]
  y = minal['경도'][i]
  folium.Marker([x, y], popup=minal['이름'][i], tooltip=minal['이름'][i], icon=folium.Icon(color='red', icon='star')).add_to(map)
```
