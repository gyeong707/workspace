# 2. Random Sampling (without replacement)

```
len(data) #데이터 인덱스와 같음.
```

```
nums = []
ran_num = random.randint(0, len(data)) #0~835255 사이의 값 랜덤샘플링

for i in range(6000):
  while ran_num in nums:
      ran_num = random.randint(0, len(data))
  nums.append(ran_num)

print(len(nums))
```

```
nums_Ellie = nums[:1000]
nums_Anna = nums[1000:2000]
nums_Kio = nums[2000:3000]
nums_Sun = nums[3000:4000]
nums_Jane = nums[4000:5000]
nums_Julia = nums[5000:6000]
```

```
print(len(nums_Ellie))
print(len(nums_Anna))
print(len(nums_Kio))
print(len(nums_Sun))
print(len(nums_Jane))
print(len(nums_Julia))
```

```
label_Ellie = data.loc[nums_Ellie]
label_Anna = data.loc[nums_Anna]
label_Kio = data.loc[nums_Kio]
label_Sun = data.loc[nums_Sun]
label_Jane = data.loc[nums_Jane]
label_Julia = data.loc[nums_Julia]
```

```
print(len(label_Ellie))
print(len(label_Anna))
print(len(label_Kio))
print(len(label_Sun))
print(len(label_Jane))
print(len(label_Julia))
```

```
#현재 인덱스 초기화
label_Ellie.reset_index(drop=True, inplace=True)
label_Anna.reset_index(drop=True, inplace=True)
label_Kio.reset_index(drop=True, inplace=True)
label_Sun.reset_index(drop=True, inplace=True)
label_Jane.reset_index(drop=True, inplace=True)
label_Julia.reset_index(drop=True, inplace=True)
```

```
#데이터셋 각각 저장
def save(data, i):
  name = ['Ellie', 'Anna', 'Kio', 'Sun', 'Jane', 'Julia']
  data.to_csv('label_'+name[i]+'.csv', index_label = 'idx2')
```

```
save(label_Ellie, 0)
save(label_Anna, 1)
save(label_Kio, 2)
save(label_Sun, 3)
save(label_Jane, 4)
save(label_Julia, 5)
```
