# Wikipida Crawler (info-table)
위키피디아에 프로그램 이름을 검색하면 나오는, 표로 구성된 정보를 긁어오는 크롤러

```python
import time
from selenium import webdriver
from selenium.common.exceptions import NoSuchElementException,StaleElementReferenceException
import os
import pandas as pd
```

## 1. Data load

```python
drama = pd.read_csv("크롤링용_모집단.csv")
```

```python
drama
```

## 2. Wikipida info-table cralwer

```python
name = drama['name']
country = []
channel = []
period = []
time = []
count = []
max_rating = []
avg_rating = []

for i in range(len(name)):
    country.append("") 
    channel.append("")
    period.append("")
    time.append("")
    count.append("")
    max_rating.append("")
    avg_rating.append("")
    
# country = drama['country']
# channel = drama['channel']
# period = drama['period']
# count = drama['count']
# max_rating = drama['max_rating']
# avg_rating = drama['avg_rating']
```

```python
print(len(name), len(country), len(channel), len(period), len(count), len(max_rating), len(avg_rating))
```

```python
# driver
driver = webdriver.Chrome(executable_path="C:\chromedriver\chromedriver.exe")

for i in range(1,741):
    driver.get("https://ko.wikipedia.org/wiki/%EC%9C%84%ED%82%A4%EB%B0%B1%EA%B3%BC:%EB%8C%80%EB%AC%B8")
    elem = driver.find_element_by_id('searchInput')
    elem.send_keys(name[i])
    elem.submit()
    
    ##### crawling #####
    print("-------" + str(i) + "행을 크롤링중입니다.--------")
    print("< " + name[i] + " >")
    
    try:
        table = driver.find_element_by_tag_name('table')
        tbody = table.find_element_by_tag_name("tbody")
        trs = tbody.find_elements_by_tag_name("tr")
    
    except NoSuchElementException:
        print("Nothing")
        continue
        
    for tr in trs:
        if "방송 국가" in tr.text:
            td = tr.find_element_by_tag_name("td")
            print(td.text)
            country[i] = td.text
            
        if "방송 채널" in tr.text:
            td = tr.find_element_by_tag_name("td")
            print(td.text)
            channel[i] = td.text
            
        if "방송 기간" in tr.text:
            td = tr.find_element_by_tag_name("td")
            print(td.text)
            period[i] = td.text
             
        if "방송 시간" in tr.text:
            td = tr.find_element_by_tag_name("td")
            print(td.text)
            time[i] = td.text
            
        if "방송 횟수" in tr.text:
            td = tr.find_element_by_tag_name("td")
            print(td.text)
            count[i] = td.text
        
print("This is result")
print("Do all list have same length?")
print(len(country) == len(channel) == len(period) == len(time) == len(count))
print(len(country), len(channel), len(period), len(time), len(count))
```

## 3. Data save

```python
drama['country'] = pd.DataFrame(country)
drama['channel'] = pd.DataFrame(channel)
drama['period'] = pd.DataFrame(period)
drama['time'] = pd.DataFrame(time)
drama['count'] = pd.DataFrame(count)
```

# Nan인 애들 드라마이름 + (드라마) 추가하여 크롤링
결측치가 많아서 확인해보았더니 도깨비와 같이 드라마 이름에 해당되는 다른 단어가 있는 경우 도깨비 (드라마)로 검색해야 결과가 나오는 것을 알 수 있었다.  
이런 행들을 위해 결측값인 데이터들에 (드라마)를 추가하여 다시 크롤링해보자.

## 1. Data load

```python
drama2 = pd.read_csv("info_crawling.csv")
```

```python
drama2 = drama2.sort_values("country")
```

```python
drama2
```

```python
drama_notnull = drama2[drama2['country'].notnull()]
```

```python
drama_notnull
```

```python
drama_null = drama2[drama2['country'].isnull()]
```

```python
drama_null.reset_index(inplace=True)
```

## 2. Cralwer (sendkey 보낼 때 이름만 바꾼 것 빼고는 똑같다)

```python
name = drama_null['name']
country = []
channel = []
period = []
time = []
count = []
max_rating = []
avg_rating = []

for i in range(len(name)):
    country.append("") 
    channel.append("")
    period.append("")
    time.append("")
    count.append("")
    max_rating.append("")
    avg_rating.append("")
```

```python
# driver
driver = webdriver.Chrome(executable_path="C:\chromedriver\chromedriver.exe")

for i in range(0,len(name)):
    driver.get("https://ko.wikipedia.org/wiki/%EC%9C%84%ED%82%A4%EB%B0%B1%EA%B3%BC:%EB%8C%80%EB%AC%B8")
    elem = driver.find_element_by_id('searchInput')
    elem.send_keys(name[i] + " (드라마)")
    elem.submit()
    
    ##### crawling #####
    print("-------" + str(i) + "행을 크롤링중입니다.--------")
    print("< " + name[i] + " >")
    
    try:
        table = driver.find_element_by_tag_name('table')
        tbody = table.find_element_by_tag_name("tbody")
        trs = tbody.find_elements_by_tag_name("tr")
    
    except NoSuchElementException:
        print("Nothing")
        continue
        
    for tr in trs:
        if "방송 국가" in tr.text:
            td = tr.find_element_by_tag_name("td")
            print(td.text)
            country[i] = td.text
            
        if "방송 채널" in tr.text:
            td = tr.find_element_by_tag_name("td")
            print(td.text)
            channel[i] = td.text
            
        if "방송 기간" in tr.text:
            td = tr.find_element_by_tag_name("td")
            print(td.text)
            period[i] = td.text
             
        if "방송 시간" in tr.text:
            td = tr.find_element_by_tag_name("td")
            print(td.text)
            time[i] = td.text
            
        if "방송 횟수" in tr.text:
            td = tr.find_element_by_tag_name("td")
            print(td.text)
            count[i] = td.text
        
print("This is result")
print("Do all list have same length?")
print(len(country) == len(channel) == len(period) == len(time) == len(count))
print(len(country), len(channel), len(period), len(time), len(count))
```

```python
drama_null['country'] = pd.DataFrame(country)
drama_null['channel'] = pd.DataFrame(channel)
drama_null['period'] = pd.DataFrame(period)
drama_null['time'] = pd.DataFrame(time)
drama_null['count'] = pd.DataFrame(count)
```

```python
drama_null
```

## 3. Save Data

```python
drama3 = pd.concat([drama_notnull, drama_null])
```

```python
drama3 = drama3.reset_index()
```

```python
drama3.to_csv("info_crawling_2.csv", encoding='utf-8-sig')
```
